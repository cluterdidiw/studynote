```python
mlb1 = dataWoo('mlb1')
```

```python
mlb1.info()
```

```python
x = mlb1['gamesyr','rbisyr','hrunsyr']
y = np.log(mlb1['salary'])
```

```python
x = sm.add_constant(x) # 若模型中有截距，必须有这一步
model = sm.OLS(y, x).fit() # 构建最小二乘模型并拟合
print(model.summary()) # 输出回归结果
```

### 大学GPA的决定因素例子

```python
GPA1 = dataWoo('gpa1')
```

```python
GPA1.info()
```

```python
GPA1.describe().T
```

```python
x = GPA1['hsGPA']
y = GPA1['colGPA']
```

```python
sns.jointplot(x = 'hsGPA',y = 'colGPA',data = GPA1,kind = 'reg')

x = sm.add_constant(x) # 若模型中有截距，必须有这一步
model = sm.OLS(y, x).fit() # 构建最小二乘模型并拟合
print(model.summary()) # 输出回归结果
```

```python
x = GPA1[['hsGPA','ACT']]
y = GPA1['colGPA']
cols = ['colGPA','hsGPA','ACT']
```

```python
sns.pairplot(data=GPA1[cols],height=2.5)

sns.jointplot(x = 'hsGPA',y = 'colGPA',data = GPA1,kind = 'reg')
sns.jointplot(x = 'ACT',y = 'colGPA',data = GPA1,kind = 'reg')
sns.jointplot(x = 'hsGPA',y = 'ACT',data = GPA1,kind = 'reg')


x = sm.add_constant(x) # 若模型中有截距，必须有这一步
model = sm.OLS(y, x).fit() # 构建最小二乘模型并拟合
print(model.summary()) # 输出回归结果
```
